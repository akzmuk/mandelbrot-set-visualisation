# Mandelbrot set visualisation with Matplotlib library

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iters=100):
    z = c
    for n in range(max_iters):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iters

width, height = 4, 4
re_start, re_end = -2, 1
im_start, im_end = -1.5, 1.5

re = np.linspace(re_start, re_end, width * 100)
im = np.linspace(im_start, im_end, height * 100)

mandel = np.zeros((height * 100, width * 100))
for i, x in enumerate(re):
    for j, y in enumerate(im):
        mandel[j, i] = mandelbrot(x + 1j * y)

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(mandel, cmap='binary', extent=(re_start, re_end, im_start, im_end))
ax.set_xlabel('Re')
ax.set_ylabel('Im')
plt.show()
```
